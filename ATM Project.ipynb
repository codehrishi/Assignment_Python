#                              ATM Simulator Project

```python
class ATM:
    def __init__(self, account_number, pin, balance):
        self.account_number = account_number
        self.pin = pin
        self.balance = balance

    def check_balance(self):
        return self.balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            return f"Deposited ${amount:.2f}. Current Balance: ${self.balance:.2f}"
        else:
            return "Invalid deposit amount."

    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            return f"Withdrew ${amount:.2f}. Current Balance: ${self.balance:.2f}"
        else:
            return "Invalid withdrawal amount or insufficient funds."

def main():
    # Example account with a predefined PIN
    account_number = "1234567890"
    pin = "1234"
    balance = 1000.0

    atm = ATM(account_number, pin, balance)

    # Prompt the user for PIN
    user_pin = input("Enter your PIN: ")
    if user_pin != pin:
        print("Incorrect PIN. Exiting.")
        return

    while True:
        print("ATM Simulator")
        print("1. Check Balance")
        print("2. Deposit")
        print("3. Withdraw")
        print("4. Exit")

        choice = input("Select an option: ")

        if choice == "1":
            print(f"Current Balance: ${atm.check_balance():.2f}")
        elif choice == "2":
            amount = float(input("Enter deposit amount: $"))
            result = atm.deposit(amount)
            print(result)
        elif choice == "3":
            amount = float(input("Enter withdrawal amount: $"))
            result = atm.withdraw(amount)
            print(result)
        elif choice == "4":
            print("Exiting ATM.")
            break
        else:
            print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()
```
